```python
def positive_feedback(ratings):
    if not ratings:
        return 0

    positive = [r for r in ratings if r >= 4]
    percentage = (len(positive) / len(ratings)) * 100
    return percentage


ratings = [5, 4, 3, 5, 2, 4, 1, 5]

print("Positive Feedback:", positive_feedback(ratings), "%")
```
