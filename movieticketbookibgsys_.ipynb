```python
def book_seat(booked, seat, total):
    if seat not in booked and 1 <= seat <= total:
        booked.append(seat)

def cancel_seat(booked, seat):
    if seat in booked:
        booked.remove(seat)


total_seats = 10
booked_seats = [2, 5, 7]

book_seat(booked_seats, 3, total_seats)
cancel_seat(booked_seats, 5)

available = [i for i in range(1, total_seats + 1) if i not in booked_seats]

print("Available seats:", available)
```
