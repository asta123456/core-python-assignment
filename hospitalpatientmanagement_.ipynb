```python
def search_patient(patients, disease):
    names = []
    for p in patients:
        if p["Disease"] == disease:
            names.append(p["Name"])
    return names


patients = [
    {"Name": "Alice", "Age": 30, "Disease": "Flu"},
    {"Name": "Bob", "Age": 45, "Disease": "Diabetes"},
    {"Name": "Charlie", "Age": 35, "Disease": "Flu"}
]

print("Patients with Flu:", search_patient(patients, "Flu"))
```
